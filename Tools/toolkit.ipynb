##Toolkits

a toolkit is just a collection of related tools that serve a common purpose packaged together for convienience and reusability

```python
from langchain_core.tools import tool

@tool
def add(a:int,b: int)->int:
  '''add two numbers'''
  return a+b

@tool
def multiply(a:int,b:int)-> int:
  '''multiply two numbers'''
  return a*b
```

```python
class MathToolkit:
  def get_tools(self):
    return [add,multiply]
```

```python
toolkit=MathToolkit()
tools=toolkit.get_tools()

for tool in tools:
  print(tool.name,"=>",tool.description)
```
